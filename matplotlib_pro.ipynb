## Check installed fonts

```python
#from: http://jonathansoma.com/lede/data-studio/matplotlib/list-all-fonts-available-in-matplotlib-plus-samples/

import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])
HTML("<div style='column-count: 2;'>{}</div>".format(code))

```

## Do a test plot

```python
#the new fonts to be used
title_font = ""
axis_font = ""
label_font = ""

#generate some data
random_array1 = np.random.normal(4, size=(1,10000))
outlier_noise = np.random.normal(4,0.06,size=(1,10)) #some noise to cause outliers at a specific point
generic_noise = np.random.normal(size=(1,10000))/5 #noise applied to x axis to make data look more realistic

#plot it
figure(num=None, figsize=(14, 8), dpi=80, facecolor='w', edgecolor='k')
#plt.arrow(2, 7, 1.8, -0.9, head_width=0.1, head_length=0.15)
plt.scatter(random_array1 + generic_noise , random_array1, color='salmon', alpha=0.1)
plt.scatter(outlier_noise ,outlier_noise + 2, color='salmon', alpha=0.17)
plt.plot([-1,9],[-1,9], color='slategray' )

plt.xlim(0, 9)
plt.ylim(0, 9)

plt.title("A random plot to test fonts\n",fontname="LiberationSansNarrow-Regular")
plt.xlabel("\nTest Axis")
plt.ylabel("Test Axis\n")
```
